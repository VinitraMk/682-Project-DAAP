```python
#clean up yolo label
%cd /content/drive/My\ Drive/UMass\ Projects/CS682
import os
train_path = 'patched-images/train/labels'
valid_path = 'patched-images/val/labels'

train_fs = [fname for fname in os.listdir(train_path) if '.txt' in fname]
val_fs = [fname for fname in os.listdir(valid_path) if '.txt' in fname]

for fn in train_fs:
    fpath = os.path.join(train_path, fn)
    data = ''
    with open(fpath, 'r', encoding='utf-8') as fp:
        data = fp.readlines()[0]
    if "," in data:
        data = data.split(",")
        xc, yc, w, h = int(data[1]), int(data[2]), int(data[3]), int(data[4])
        xc = ((xc + 50) - xc / 2)/224
        yc = ((yc + 50) - yc / 2)/224
        data[1] = xc
        data[2] = yc
        data[3] = w / 224
        data[4] = h / 224
        new_ln = f'{data[0]} {data[1]} {data[2]} {data[3]} {data[4]}'

        with open(fpath, 'w+', encoding='utf-8') as fp:
            fp.writelines(new_ln)

for fn in val_fs:
    fpath = os.path.join(valid_path, fn)
    data = ''
    with open(fpath, 'r', encoding='utf-8') as fp:
        data = fp.readlines()[0]
    if "," in data:
        data = data.split(",")
        xc, yc, w, h = int(data[1]), int(data[2]), int(data[3]), int(data[4])
        xc = ((xc + 50) - xc / 2)/224
        yc = ((yc + 50) - yc / 2)/224
        data[1] = xc
        data[2] = yc
        data[3] = w / 224

        data[4] = h / 224
        new_ln = f'{data[0]} {data[1]} {data[2]} {data[3]} {data[4]}'

        with open(fpath, 'w+', encoding='utf-8') as fp:
            fp.writelines(new_ln)
```

```python
import pandas as pd
import os
from skimage import io
from torchvision.io import write_png
import torch

%cd /content/drive/My\ Drive/UMass\ Projects/CS682/yolov5

txt_path = 'runs/detect/exp19/labels'
txt_files = os.listdir(txt_path)
imgs_path = 'data/patched-images/val/images'
zeroout_imgpath = 'data/zeroout'

for fn in txt_files:
    data = []
    with open(os.path.join(txt_path, fn), 'r', encoding='utf-8') as fp:
        data = fp.readlines()[0].split(" ")
    xmin, ymin, xmax, ymax = int(data[1]), int(data[2]), int(data[3]), int(data[4])
    imgname = fn.replace(".txt", ".JPEG")
    img = io.imread(os.path.join(imgs_path, imgname))
    img[ymin:ymax, xmin:xmax, :] = 1
    img = img.transpose(2, 0, 1)
    #print(img, zeroout_imgpath, imgname)
    write_png(torch.from_numpy(img), os.path.join(zeroout_imgpath, imgname))
```

```python
%cd /content/drive/My\ Drive/UMass\ Projects/CS682
!ls
import os
import pandas as pd
data_dir = 'yolov5/data/zeroout-sm'
zeroout_fs = os.listdir(data_dir)
df = pd.read_csv('data/imagenette2/val_images.csv')
y = [df[df['filename'] == fname]['label'].values[0] for fname in zeroout_fs]
classes = os.listdir('data/imagenette2/train')
classes_id = [0, 491, 482, 217, 574, 497, 701, 513, 569, 571]


%cd /content/drive/My\ Drive/UMass\ Projects/CS682
from models.ots_models import get_model
from transforms import ToTensor, Normalize
from torchvision import transforms
from skimage import io

data_transforms = transforms.Compose([ToTensor(), Normalize()])
models_acc = { 'resnet18': 0 , 'resnet50': 0, 'inception': 0}

normalize = transforms.Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225))
N = len(zeroout_fs)
print('No of test images:', N, '\n\n')
c = 0
for modelname in models_acc:
    c = 0
    model, mid2name = get_model(modelname)
    model.eval()
    for idx, imgfs in enumerate(zeroout_fs):
        img = io.imread(f'{data_dir}/{imgfs}')
        img_tensor = torch.from_numpy(img.transpose(2, 0, 1)).double()
        primg = normalize(img_tensor).float().unsqueeze(0)
        prediction = model(primg).squeeze(0).softmax(0)
        prediction_slice = prediction
        class_id = prediction_slice.argmax().item()
        #score = prediction_slice[class_id].item()
        #category_name = mid2name[class_id]
        #cid = classes_id.index(class_id)
        c += (y[idx] == class_id)
        #print(f"File: {imgfs}, Predicted: {category_name}, True class: {y[idx]}")
    models_acc[modelname] = (c / N) * 100
print(models_acc)
```
